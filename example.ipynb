# Color palette from Image using Gaussian Mixture

## One simple example

```python
import matplotlib.pyplot as plt
import numpy as np
import cv2

from sklearn.mixture import GaussianMixture
```

```python
PATH = './img/img1.jpg'

img = cv2.imread(PATH)[...,::-1]
img_flat = img.reshape((-1,3))
```

```python
gm = GaussianMixture(n_components=5,random_state=42)
result = gm.fit(img_flat) # calculate

print("Results")
print(result.means_) # result
```

### Original img

```python
plt.imshow(img) # original img
```

### Result Palette colors

```python
plt.imshow(result.means_[np.newaxis,:].astype(int)) # show palette
```

## Result for other images

```python
def show_palette(path:str,n_components=5):
    img = cv2.imread(path)[...,::-1]
    img_flat = img.reshape((-1,3))
    gm = GaussianMixture(n_components=n_components,random_state=42).fit(img_flat)
    m = gm.means_
    plt.imshow(img)
    plt.show()
    plt.imshow(m[np.newaxis,:].astype(int))
    plt.show()
    
    return m
```
